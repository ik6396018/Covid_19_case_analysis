```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
df.head()
```

```python
df.tail()
```

```python
df=df.iloc[:36,1:]
df
```

```python
columns = df.columns
print(columns)
```

```python
plt.figure(figsize = (20 , 20))
sns.barplot(x = df["death"], y = df["state_name"])
plt.ylabel("State_Name")
plt.xlabel("Number of deaths")
plt.title("Nomber of death by state")
plt.show()
```

```python
df1 = df.iloc[:,1:4]
df1.plot.barh(figsize = (20,20))
```

```python
data_active = df.sort_values(['active'], ascending = False)
data_active = data_active.iloc[:7,:]
data_active

```

```python
explode = [0, 0.1, 0, 0, 0, 0, 0] 

colors = ["c", "b", "y", "g", "r", "m", "brown"]
textprops = {"fontsize": 15}
wedgeprops = {"linewidth": 2, "width": 1, "edgecolor": "black"}

plt.figure(figsize=(16, 19))

fontdict = {"fontsize": 20}
title_text = "Top seven corona affected states \n Total active cases in India: {}\n Total active cases in these states: {} \n Total percentage of active cases in these states: {}%".format(
    sum(df["active"]), sum(data_active["active"]), round(sum(data_active["active"]) / sum(df["active"]), 4) * 100
)
plt.title(title_text, fontdict=fontdict)

plt.pie(
    data_active["active"],
    labels=data_active["state_name"],
    explode=explode,
    colors=colors,
    autopct="%0.2f%%",
    radius=1,
    textprops=textprops,
    pctdistance=0.6,  # Corrected typo in 'pctdistance'
    labeldistance=1.1,
    wedgeprops=wedgeprops,
    startangle=90,  # Added startangle for better clarity
    rotatelabels=False,  # Corrected typo in 'rotatelabels'
)
plt.legend(loc=2)

plt.show()
```
